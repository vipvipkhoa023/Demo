# YOLOv7 training on Bright Lighting Dataset




**Dataset Overview**:

---


* **Collection Date**: May 2023
* **Location**: Symonds Street, Auckland CBD
* **Total Images**: 516 <a href="https://universe.roboflow.com/301/cs302">
    <img src="https://app.roboflow.com/images/download-dataset-badge.svg"></img>
</a>
* **Source**: Extracted from a 17-second video at 29 frames per second.

The dataset was specifically recorded against sunbeams to simulate real-life driving experiences when blinded by intense lighting. This aims to provide a more realistic training environment for deep learning models that deal with such scenarios.

`Data Augmentation`:
To enhance the dataset and ensure robust training for the deep learning model, various augmentation techniques were applied. Resizing and further augmentations were managed using Roboflow.
An enhanced version (version 2) of 1300 images can be downloaded [here](https://app.roboflow.com/301/cs302/2).


`Annotation & Segmentation`:
The images were labeled and segmented using the Grounding Dino and Sam algorithms. For a detailed walkthrough, please take a look at the Image Annotation with SAM and Grounding Dino notebook [here](https://github.com/irinagetman1973/Urban-Vehicle-Detection-via-DL/blob/main/Image_Annotation_with_SAM_and_Grounding_Dino.ipynb)
